```python
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784')
```

```python
mnist['data']
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
X, y = mnist['data'], mnist['target']
X
```

```python
random_digit = X.iloc[4560].to_numpy()
some_random_digit = random_digit.reshape(28,28)
plt.imshow(some_random_digit, cmap=matplotlib.cm.binary, interpolation="nearest")
```

```python
x_train, x_test = X[:6000] , X[6000:7000]
y_train, y_test = y[:6000], y[6000:7000]
```

```python
import numpy as np

shuffle_index = np.random.permutation(6000)
#x_train.iloc[shuffle_index]
x_train, y_train = x_train.iloc[shuffle_index], y_train.iloc[shuffle_index]
```

```python
y_train = y_train.astype(np.int8)
y_test = y_test.astype(np.int8)

y_train_2 = (y_train == 2)
y_test_2 = (y_test == 2)
```

```python
y_test_2.to_numpy()
y_train_2.to_numpy()
```

```python
(y_train == 2).value_counts()
```

```python
(y_test_2 == 2).value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#tol is tolerance 
clf = LogisticRegression(tol=0.1)
```

```python
clf.fit(x_train, y_train_2)
```

```python
y_pred = clf.predict([random_digit])
```

```python
y_pred
```

```python
from sklearn.model_selection import cross_val_score

a = cross_val_score(clf, x_train, y_train_2, cv=3, scoring='accuracy')
```

```python
a.mean()
```
